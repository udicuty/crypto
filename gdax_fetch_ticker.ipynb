```python
end=datetime.now(tz=timezone.utc).replace(microsecond=0,second=0,minute=0)
start=end-timedelta(hours=1)
end_label=str(calendar.timegm(end.utctimetuple()))
start_label=str(calendar.timegm(start.utctimetuple()))
print(end)
print(start)
```

```python
base_url='https://api.gdax.com/products/{}/candles'
products=['LTC-USD','ETH-USD','BTC-USD']
params={'start':start,'end':end,'granularity':30}

for product in products:

    base_url='https://api.gdax.com/products/{}/candles'.format(product)
    resp=requests.get(base_url,params=params)
    
    data=json.loads(resp.content)
    dump=[ticker_to_dic(l,product) for l in data]
    
    fname='_'.join(['gdax',product,start_label,end_label])+'.json'
    key='gdax/'+fname
    s3.put_object(Bucket='udi-crypto',Key=key, Body=json.dumps(dump))
    time.sleep(1) #Important to avoid thorttling
    print(fname)
```
